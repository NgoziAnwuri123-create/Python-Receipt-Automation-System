```python
#### import random

lotto_number = random.randint(1,51)
print("We have choosen a random number between 1 and 51. Try to guess it")

while True:
    guess = input("Enter your guess(number) or type 'yes' to quit")

    if guess.lower() == "yes":
        print(f"The number to guess was {lotto_number} . better luck next time")
        break

    if guess.isdigit():
        guess = int(guess)
        
    else: 
        print("please enter a valid number. try again")
        continue 

    if guess < lotto_number :
        print(" Your guess is lower than the guess number")
    elif guess > lotto_number:
        print ("Your guess is higher than the guess number")
    else:
      print("congratulation - you have won 5 millon pounds")
      break
```
